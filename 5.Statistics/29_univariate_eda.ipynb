```python
cereal=pd.read_csv('/content/cereal.csv')
```

```python
plt.figure(figsize=(20,5))
index=1
for i in cereal.columns[3:11]:
  plt.subplot(1,8,index)
  sns.histplot(cereal[i])
  index+=1
```

With the help of a barplot, compare the manufacturers based on customer ratings

Task2:
With the help of a pie chart, find out which of the manufacturers are the largest producers of cereals

```python
cereal.groupby('mfr')['rating'].mean().plot.bar()
plt.xlabel('Manufacturers')
plt.ylabel('ratings')
plt.title('manufacturers with highest customer ratings ')
plt.show()
```
