```python
import pandas as pd
import numpy as np
```

A z-test is a hypothesis test in which the z-statistic follows a normal distribution
A z-test is a statistical test used to determine whether two population means are different when the variances are known and the sample size is large.

image.png

```python
#load the sample data
from sklearn.datasets import fetch_california_housing
cali=fetch_california_housing()
```

```python
cali
```

```python
from sklearn.utils import shuffle
x,y=shuffle(cali.data,cali.target,random_state=1)
```

```python
#dataframe
df=pd.DataFrame(cali.data,columns=cali['feature_names'])
df['target']=cali.target
df.head()
```

```python
#step 1:mean,standard deviation>>>population data, 
pop_target=df.target
pop_mean=pop_target.mean()
pop_mean
```

```python
pop_std=pop_target.std()
pop_std
```

```python
#step2 get the sample>>>100
sample=100
sample_values=df.sample(n=sample,random_state=1).target
```

```python
#sample mean and sample std
sample_mean=sample_values.mean()
```

```python
sample_std=sample_values.std()
```

```python
sample_mean,sample_std
```

```python
#z statistic
import math
den=pop_std/math.sqrt(100)
(sample_mean-pop_mean)/den
```
