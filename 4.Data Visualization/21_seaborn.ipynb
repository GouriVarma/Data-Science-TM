##data visualization using seaborn


```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('/content/StudentsPerformance.csv')
df.head()
```

```python
plt.style.use('default')
sns.displot(df['math score'],color='m')
plt.show()
```

```python
iris=sns.load_dataset('iris')
```

```python
iris.head()
```

```python
sns.set(style='whitegrid')
sns.lineplot(x='sepal_length',y='sepal_width',data=iris)
```

```python
df=sns.load_dataset('anscombe')
```

```python
df.head()
```

```python
#basic plot>>>> shows a line which represents a linear regression model along the data points>>>x and y 
sns.lmplot(x='x',y='y',data=df)
```

```python
sns.barplot(x='species',y='sepal_length',data=iris)
```

```python
list1=[33,44,22,56,67]
labels=['a','b','c','d','e']
palette=sns.color_palette('BuGn')
plt.pie(list1,labels=labels,colors=palette,autopct='%.0f%%')
plt.show()
```
